```python
#Inheritance adalah pewarisan class kepada class lain, 
#class Induk : class yang mewariskan
#class Anak : menerima warisan dri class induk
#class yang menerima warisan (class anak), bisa mengakses properti & Method pada class yang 

#mewariskan (class induk). 
#Konsep ini juga digunakan untuk menghindari duplikasi kode program
#Namun, induk classnya tak bisa mengakses properti anaknya(class yang menerimawarisan), 

#seperti halnya class anak yang mewarisi class induknya . 

#CONTOH

class user:
    def __init__(self,name):
        self.name=name
    def printname(self):
        print("Name="+ self.name)
class programer(user):#mengikuti kelas induk (user)
    def __init__(self,name):
        self.name=name
        
    def dopyhton(self):
        print("programing python")
       
class doPython(user):
    pass

user=user('brian')
user.printname()

programer=programer('diana')

programer.printname()
programer.dopyhton()
```

```python
#Inheritance (Pewarisan Properti dan method dari induk class) memiliki beberapa type, diantaranya :
#Default,Overriding,Super,Multiple
#Default : Isinya sama dengan class induknya
'''class hiu(induk):
    pass'''
#Overriding overiding method: mengubah parameter dari fungsi yang dipanggil dari kelas induk,
#methodnya sama isinya beda,
#Anak kelas yang merubah properti / Method pada kelas induk,bisa juga diartikan
#method yang parrent class yang ditulis kembali oleh subclass.
class Persegi:        
    def __init__(self,segi,bentuk="persegi"):
        self.segi=segi
        self.bentuk=bentuk
    def display(self):
        print("sisinya= ",self.segi,"bentuknya= ",self.bentuk)
    def deskripsi(self):
        print ("4 sisinya sama panjang")
        
class PersegiPanjang(Persegi):        
    def __init__(self,segi,bentuk="persegi panjang"):
        self.segi=segi
        self.bentuk=bentuk
    def deskripsi(self):
        print ("ada 2 pasang sisi yang sama panjang")
a=Persegi(4)
a.display()
a.deskripsi()
b=PersegiPanjang(4)
b.display()
b.deskripsi() 
```

```python
class Segitiga:
    def __init__(self,jumlahsegi="3",bentuksegitiga="Sembarang"):
        self.jumlahsegi=jumlahsegi
        self.bentuksegitiga=bentuksegitiga
    def ps(self):
        print(self.jumlahsegi,self.bentuksegitiga)
        
class SegitigaSamaKaki(Segitiga):
    def __init__(self,jumlahsegi="3",bentuksegitiga="Sama Kaki"):
        self.jumlahsegi=jumlahsegi
        self.bentuksegitiga=bentuksegitiga
        
class SegitigaSamaSisi(Segitiga):
    def __init__(self,jumlahsegi="3",bentuksegitiga="Sama Sisi"):
        self.jumlahsegi=jumlahsegi
        self.bentuksegitiga=bentuksegitiga

s=Segitiga()
s.ps()
a=SegitigaSamaKaki()
a.ps()
b=SegitigaSamaSisi()
b.ps()
```

```python
#Super
#Digunakan Untuk Inisiasi properti class parrent pada class anak
class Fish:
    def __init__(self,fish="Ikan"):
        self.fish=fish
    def pnj(self):
        print(self.fish)
        print(self.water)
class trout(Fish):
    def __init__(self,water="hiu"):
        self.water=water
        Fish.__init__(self) #super
        #Atau :
        #super().__init__(self)
a=trout()
a.pnj()
```

```python
#Multiple
#Anak class yang memiliki 2 atau lebih Parent class
class psn:
    def __init__(self):
        self.nama=input("Nama : ")
    def dp(self):
        print(self.nama)
class mark:
    def __init__(self):
        self.npm=input("Npm :")
        self.mtk=input("Mtk :")
        self.bio=input("Bio :")
    def dps(self):
        print("npm :", self.npm)
        print("nilai : ",int(self.mtk+self.bio)//2)
class mhs(psn,mark):#memiliki 2 parent class diatas ,class psn dan class mark
    def __init__(self):
        psn.__init__(self)
        mark.__init__(self)
    def pnk(self):
        print(psn.dp(self),
             mark.dps(self))
a=mhs()
a.pnk()
```
