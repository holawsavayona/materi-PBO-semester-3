```python
class point:
    def __init__(self,x=0,y=0):
        self.x=x
        self.y=y
    def __str__(self):
        return ("point object is at:("+str(self.x)+","+str(self.y)+")")
    def __sub__(self, other):#untuk fungsi -(hanya untuk pemanggilan fungsi)
        x=self.x-other.x
        y=self.y-other.y
        return point(x,y)
    def __add__(self, other): #untuk fungsi + (hanya untuk pemanggilan fungsi)
        x=self.x-other.x
        y=self.y-other.y
        return point(x,y)
    
    
p1=point(2,3)
p2=point(-1,2)
print (p1-p2)
print(p1.__sub__(p2))#(hanya untuk pemanggilan fungsi) dari spesial method diatas 
print (p1+p2)
print(p1.__add__(p2))#(hanya untuk pemanggilan fungsi) dari spesial method diatas
```
